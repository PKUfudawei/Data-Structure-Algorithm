```python
## to be improved in time complexity

import math
INF=math.inf

class Edge:
    def __init__(self, start, end, weight):
        self.start, self.end, self.weight = start, end, weight
        
def degree(graph):
    indegree = {v: 0 for v in graph}
    outdegree = {v: 0 for v in graph}
    for v in graph:
        for e in graph[v]:
            indegree[e.end] += 1
            outdegree[v] += 1
    return indegree, outdegree

def topoSort(graph):
    n = len(graph)
    indegree, outdegree = degree(graph)
    frontier = [v for v in graph if indegree[v]==0]
    sequence = []
    
    while frontier:
        v = frontier.pop(0)
        sequence.append(v)
        for e in graph[v]:
            indegree[e.end] -= 1
            if indegree[e.end] == 0:
                frontier.append(e.end)
            
    if len(sequence) != n:
        return None
    return sequence

def AOE(graph):
    indegree, outdegree=degree(graph)
    topoList=topoSort(graph)
    earliest, latest = {v: -INF for v in graph}, {v: +INF for v in graph}
    
    for j in topoList:
        if indegree[j]==0:
            earliest[j] = 0
        else:
            earliest[j]=max(earliest[i]+e.weight for i in topoList for e in graph[i] if e.end==j)
            
    T=max(earliest.values())
    
    for i in topoList[::-1]:
        if outdegree[i]==0:
            latest[i] = T
        else:
            latest[i]=min(latest[j]-e.weight for j in topoList for e in graph[i] if e.end==j)
    
    return T, earliest, latest


def driver():
    n, m = map(int, input().split())
    graph={}
    edges=set()
    weights={}
    for _ in range(m):
        start, end, weight = map(int, input().split())
        graph[start] = graph.get(start, []) + [Edge(start, end, weight)]
        weights[(start, end)] = weight
        if end not in graph:
            graph[end]=[]
        edges.add((start, end))
        
    shortest, earliest, latest = AOE(graph)
    print(shortest)
    keyEvent=set()
    for i in earliest:
        if earliest[i]==latest[i]:
            keyEvent.add(i)
            
    keyAction=[]      
    for i in keyEvent:
        for j in keyEvent:
            if (i,j) in edges and weights[(i,j)]==earliest[j]-earliest[i]:
                keyAction.append((i,j))
    keyAction.sort()
    for i,j in keyAction:
        print(i,j)
        
driver()
```
