```python
class Stack:
    def __init__(self) -> None:
        self.items=[]
        self.maxItem=[]
        self.minItem=[]
    
    def push(self, item):
        self.items.append(item)
        if len(self.maxItem)==0:
            self.maxItem.append(item)
        elif self.maxItem[-1]<item:
            self.maxItem.append(item)
        else:
            self.maxItem.append(self.maxItem[-1])
            
        if len(self.minItem)==0:
            self.minItem.append(item)
        elif self.minItem[-1]>item:
            self.minItem.append(item)
        else:
            self.minItem.append(self.minItem[-1])
    
    def pop(self):
        if len(self.items)==0:
            return
        self.maxItem.pop()
        self.minItem.pop()
        return self.items.pop()
        
    def top(self):
        return self.items[-1]

    def isEmpty(self):
        return len(self.items)==0

    def max(self):
        if len(self.maxItem)==0:
            return
        else:
            print(self.maxItem[-1])

    def min(self):
        if len(self.minItem)==0:
            return
        else:
            print(self.minItem[-1])
    
def commands(stack, string):
    if 'pop' in string:
        stack.pop()
    elif 'push' in string:
        stack.push(int(string.split()[-1]))
    elif 'min' in string:
        return stack.min()
    
def driver():
    pigs=Stack()
    try:
        while True:
            commands(pigs,input())
    except EOFError:
        pass
        
driver()
```
