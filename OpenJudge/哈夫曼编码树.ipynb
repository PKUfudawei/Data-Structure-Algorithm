```python
class BinaryTree:
    def __init__(self, data, left=None, right=None):
        self.data=data
        self.left=left
        self.right=right
        

def specialMin(trees):
    Min = None
    Minkey = None
    for key in trees.keys():
        if Min is None:
            Min=trees[key].data
            Minkey=key
        else:
            if trees[key].data<Min:
                Min=trees[key].data
                Minkey=key
            elif trees[key].data==Min:
                if min(Minkey)>min(key):
                    Min=trees[key].data
                    Minkey=key
    return Minkey
                   
        
def huffmanTreeEncoder(tuples, trees={}, codes={}):
    if len(trees)==0:
        for c,freq in tuples:
            trees[c]=BinaryTree(freq)
            codes[c]=''
    
            
    key1=specialMin(trees)
    tree1=trees.pop(key1)
    key2=specialMin(trees)
    tree2=trees.pop(key2)
            
    for c in key1:
        codes[c]='0'+codes[c]
    for c in key2:
        codes[c]='1'+codes[c]
                
    trees[key1+key2]=BinaryTree(tree1.data+tree2.data, left=tree1, right=tree2)
    
    if len(trees)==1:
        return codes
    else:
        return huffmanTreeEncoder(tuples, trees=trees, codes=codes)
    
def printCodes(string, codes):
    for c in string:
        print(codes[c], end='')
    
def decode(string, codes):
    decodes={v:k for k,v in codes.items()}
    i=0
    while i < len(string):
        for j in range(i, len(string)):
            if string[i:j+1] in decodes:
                print(decodes[string[i:j+1]], end='')
                i=j+1
                break
          
    
def driver():
    n=int(input())
    tuples=[]
    for _ in range(n):
        c, freq=input().split()
        freq=int(freq)
        tuples.append(tuple([c, freq]))
        
    codes=huffmanTreeEncoder(tuples, trees={}, codes={})
    
    while True:
        try:
            string=input()
            if string.isdigit():
                decode(string, codes)
                print()
            else:
                printCodes(string, codes)
                print()
        except:
            break

driver()
```
